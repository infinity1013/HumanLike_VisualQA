```python
import pandas as pd
from pandas import DataFrame
import json
with open('E:\Download\VisualMRC\Code\data\\train.jsonl', 'r') as json_file:
    lines = json_file.readlines()
```

```python
list=[]
for line in lines:
    json_line=json.loads(line.strip())
    l=len(json_line['bounding_boxes']) 
    for n_roi in range(0,l):
        l1=[]
        l1.append(json_line['image_filename'])
        l1.append(json_line['bounding_boxes'][n_roi]['shape']['x'])
        l1.append(json_line['bounding_boxes'][n_roi]['shape']['y'])
        l1.append((json_line['bounding_boxes'][n_roi]['shape']['width']+json_line['bounding_boxes'][n_roi]['shape']['x']))
        l1.append((json_line['bounding_boxes'][n_roi]['shape']['height']+json_line['bounding_boxes'][n_roi]['shape']['y']))
        l1.append(json_line['bounding_boxes'][n_roi]['structure'])
        list.append(l1)
    
# Create the pandas DataFrame
df = pd.DataFrame(list, columns = ['image_filename','x_min','y_min','x_max','y_max','label'])
  
# print dataframe.
df.head(10)
```

```python
import cv2
import pytesseract

pytesseract.pytesseract.tesseract_cmd = r"E:\Program Files\Tesseract-OCR\tesseract.exe"

image = cv2.imread(df['image_filename'][50], 0)
thresh = 255 - cv2.threshold(image, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]

x_min = df['x_min'][50]
x_max = df['x_max'][50]
y_min = df['y_min'][50]
y_max = df['y_max'][50]
ROI = thresh[y_min:y_max,x_min:x_max]
data = pytesseract.image_to_string(ROI, lang='eng',config='--psm 6')
print(data)

cv2.imshow('thresh', thresh)
cv2.imshow('ROI', ROI)
cv2.waitKey()
```
